```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
wine=pd.read_csv("wine.csv")
wine
```

```python
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
df_norm=norm_func(wine.iloc[:,1:])
df_norm
```

```python
from sklearn.preprocessing import MinMaxScaler
trans=MinMaxScaler()
data=pd.DataFrame(trans.fit_transform(wine.iloc[:,1:]))
data
```

```python
from scipy.cluster.hierarchy import linkage
import scipy.cluster.hierarchy as sch
z = linkage(df_norm, method="complete",metric="euclidean")
plt.figure(figsize=(15, 5))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Index')
plt.ylabel('Distance')
sch.dendrogram(
    z,
    #leaf_rotation=0.,  # rotates the x axis labels
    #leaf_font_size=8.,  # font size for the x axis labels
)
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
h_complete=AgglomerativeClustering(n_clusters=5,linkage="complete",affinity="euclidean").fit(df_norm)

cluster_labels=pd.Series(h_complete.labels_)
cluster_labels
wine['clust']=cluster_labels
wine
```

```python
wine.iloc[:,1:].groupby(wine.clust).mean()
```

```python
#NOW USING K-MEANS CLUSTERING
```

```python
from sklearn.cluster import KMeans
fig=plt.figure(figsize=(10,8))
WCSS=[]
for i in range(1,8):
    clf=KMeans(n_clusters=i)
    clf.fit(df_norm)
    WCSS.append(clf.inertia_)
plt.plot(range(1,8),WCSS)
plt.title("THE ELBOW CURVE")
plt.ylabel("WCSS")
plt.xlabel("Number of clusters")
plt.show()
```

```python
clf=KMeans(n_clusters=5)
y_kmeans=clf.fit(df_norm)
```

```python
y_kmeans
clf.labels_
```

```python
xy=pd.Series(y_kmeans)
wine['clust']=xy
wine
```

```python
wine.iloc[:,1:].groupby(wine.clust).mean()
```

```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale
import matplotlib.pyplot as plt
```

```python
wine=pd.read_csv("wine.csv")
wine.describe()
wine.head()
```

```python
wine.data=wine.iloc[:,1:]
wine.data.head()

W=wine.data.values
W
```

```python
wine_normal=scale(W)
wine_normal
```

```python
pca=PCA(n_components=13)
pca_values=pca.fit_transform(wine_normal)
pca_values
```

```python
pca.components_
```

```python
var=pca.explained_variance_ratio_
var
```

```python
var1=np.cumsum(np.round(var,decimals=4)*100)
var1
```

```python
plt.plot(var1,color='blue')
```

```python
pca_values[:,0:9]
```

```python
finalDf=pd.concat([pd.DataFrame(pca_values[:,0:9],columns=['pc1','pc2','pc3','pc4','pc5','pc6','pc7','pc8','pc9']),wine[['Type']]],axis=1)
finalDf
```

```python
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
new_norm=norm_func(finalDf.iloc[:,0:8])
new_norm
```

```python
from scipy.cluster.hierarchy import linkage
import scipy.cluster.hierarchy as sch
z = linkage(new_norm, method="complete",metric="euclidean")
plt.figure(figsize=(15, 5))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Index')
plt.ylabel('Distance')
sch.dendrogram(
    z,
    #leaf_rotation=0.,  # rotates the x axis labels
    #leaf_font_size=8.,  # font size for the x axis labels
)
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
h_complete=AgglomerativeClustering(n_clusters=5,linkage="complete",affinity="euclidean").fit(new_norm)

cluster_labels=pd.Series(h_complete.labels_)
cluster_labels
finalDf['clust']=cluster_labels
finalDf
```

```python
finalDf.iloc[:,0:].groupby(finalDf.clust).mean()
```
